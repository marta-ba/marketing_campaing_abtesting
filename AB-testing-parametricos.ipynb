# Conceptos previos

- **Tamaño muestral**: Se refiere al número de observaciones que se incluyen en un estudio o experimento. En el contexto del un prueba A/B, el tamaño muestral es fundamental porque afecta la precisión y la fiabilidad de los resultados obtenidos. 

- **Hipótesis nula y alternativa**: Estos conceptos surgen como resultado de una determinada investigación realizada sobre una muestra o una población. La hipótesis sirve como primer paso en la investigación, aunque se plantee con menor o mayor datos reales y se buscará demostrar la teoría o refutarla. La hipótesis alternativa sostiene la conclusión que el investigador quiere afirmar o demostrar de su estudio, y es su hipótesis principal. Esta teoría se identifica con la expresión **$ H_1 $**. La hipótesis alternativa es una conjetura que intenta demostrar la falsedad de la hipótesis nula, formulada en una investigación y/o un experimento.  Por otro lado, la hipótesis nula, identificada como **$ H_0 $**, es una declaración que asume que no hay diferencia o efecto significativo en las variables que se están comparando en un experimento. En otras palabras, establece que cualquier cambio observado en los datos es el resultado del azar o del error experimental, en lugar de una verdadera diferencia o efecto entre las condiciones experimentales.

    Por ejemplo, en un experimento de A/B testing, la **$ H_0 $** podría afirmar que no hay diferencia en las tasas de clics entre dos versiones de un anuncio, mientras que la **$ H_1 $** podría afirmar que una versión tiene una tasa de clics significativamente mayor que la otra.


- **P-value**: Evalúa y mide qué tan compatibles son los datos con la hipótesis nula. Es decir, contesta a la siguiente pregunta ¿Qué tan probable es el efecto observado en los datos si la **$ H_0 $** es verdadera?

    <div style="background-color: #F74646; padding: 10px; border-left: 6px solid #000080; color: black; border-radius: 10px;">
    <strong>En este contexto p-value se define como la probabilidad de rechazar la hipótesis nula.</strong>

    Se recomienda rechazar la hipótesis nula cuando el p-value es inferior a 0,05, lo que significa que existe una diferencia significativa entre Variación A y Variación B. Por el contrario si el p-value es mayor a 0,05, no podemos concluir que existe una diferencia significativa entre las variaciones.
    </div>



- **Error tipo I y tipo II**: Recordemos que estos errores pueden surgir debido a la naturaleza aleatoria de los datos, al tamaño de la muestra utilizado en el experimento, a la elección del nivel de significancia, así como a la sensibilidad del test estadístico utilizado para realizar las pruebas de hipótesis. La minimización de estos errores y la maximización de la potencia estadística son objetivos importantes en el diseño y análisis de experimentos.

    - El error tipo I, también llamado alfa (α), se comete al rechazar la hipótesis nula (**$ H_0 $**) siendo verdadera.

    - El error tipo II, o beta (β), se comete al aceptar la hipótesis nula (**$ H_0 $**) siendo falsa. 


# Introducción a A/B Testing

## Conceptos básicos de A/B testing y su importancia en la toma de decisiones basada en datos.

El A/B testing, también conocido como prueba dividida (*split testing*), es un método para comparar dos versiones diferentes de un diseño (una página web o una interfaz de usuario de una aplicación) para determinar cuál funciona mejor. Pensémoslo de esta manera. Tenemos dos variaciones de nuestro diseño, y ambas se ven bien. De forma subjetiva, es posible que prefiramos una sobre la otra. Sin embargo, queremos tomar una decisión sobre cuál elegir basada en datos. Entonces dividimos aleatoriamente a nuestro público en dos grupos (generalmente 50/50). Las personas dentro del primer grupo verán la primera variante del diseño (A), que llamamos control, y las personas del segundo grupo verán otra variante del diseño (B), que llamamos test. Lo que hace que esta prueba sea excelente es que las empresas obtienen comentarios directos de sus usuarios reales al presentarles las opciones de producto/características existentes frente a las variaciones, y de esta manera pueden probar rápidamente nuevas ideas y si estas funcionan. En caso de que la prueba A/B muestre que el test no es efectivo, al menos las empresas podrán aprender de ello y decidir si necesitan mejorarlo o si necesitan buscar otras ideas.


Si entendieramos este tipo de test como una pelea de boxeo, para tomar prestada una analogía del mundo del deporte, el control sería el "campeón", el titular actual que ha ganado previas peleas y defendido su título contra anteriores luchadores. Por otro lado, el test sería el nuevo "luchador", el boxeador compitiendo contra el campeón para ganar el título. Así que aquí lo tenemos, dos diseños luchando por nuestra atención, cada uno tratando de superar al otro.

Otro punto importante es que no hay una sola solución para todos los casos. Si bien las pruebas A/B son una gran herramienta para diseñadores de experiencia de usuario y especialistas en marketing (entre otros), es importante comprender sus limitaciones. Las pruebas A/B no son adecuadas para todas las situaciones. No es el mejor enfoque cuando se trata de múltiples cambios en el diseño, rediseños importantes o al agregar nuevas características a nuestras páginas y aplicaciones. En esos casos, podríamos obtener mejores resultados utilizando otros métodos como pruebas multivariables.

LLegados a este punto, nos podríamos preguntar por qué y cuándo realizar pruebas de A/B testing.  Las pruebas A/B **son más efectivas** cuando probamos un cambio concreto y específico (1-2 variables a la vez). En otras palabras, es una herramienta perfecta cuando tenemos dudas sobre una decisión de diseño que tiene dos alternativas claras y diferentes. Elementos como *call to action* (CTA), esquemas de color, diseño de página, titulares, descripciones de productos son generalmente ideales para pruebas A/B. Por otro lado, **no debemos utilizar** las pruebas A/B para experimentar con cambios complejos o múltiples. En conclusión, las pruebas A/B son principalmente una herramienta de optimización, no una solución rápida para problemas importantes de experiencia de usuario (UX).

**Ventajas de la prueba A/B**

- Permite aprender qué funciona y qué no de manera rápida.

- Obtenemos comentarios directamente de los clientes/usuarios del producto.

- Dado que los usuarios no son conscientes de que están siendo probados, los resultados serán imparciales.



**Desventajas de la prueba A/B**

- Presentar contenido/precio/características diferentes a diferentes clientes, especialmente en la misma geolocalización, podría ser potencialmente peligroso y resultar en "Aversión al Cambio" (discutiremos cómo abordar esto más adelante).

- Requiere una cantidad significativa de recursos de producto, ingeniería y gestión de los datos.

- Podría llevar a conclusiones incorrectas si no se lleva a cabo correctamente.

**Veamos un ejemplo práctico para entender mejor qué es la prueba A/B.**

Imagina que eres trabajador del mundo de las finanzas y decidiste lanzar una campaña publicitaria para promocionar los servicios de tu banco. El objetivo es atraer más clientes y aumentar la conversión de visitantes del sitio web en usuarios activos del banco. Sin embargo, con el tiempo, te das cuenta de que la campaña no está teniendo el impacto esperado. Aunque las visitas al sitio web aumentan, la tasa de conversión, es decir, la proporción de visitantes que se convierten en clientes, está disminuyendo gradualmente. No estás seguro de cuál es la razón detrás de esto, así que decides realizar cambios en la campaña, manteniendo el mismo objetivo y contenido, para ver si puedes revertir la tendencia.

Para realizar este cambio de campaña de manera controlada y asegurarte de que la nueva versión sea efectiva, decides utilizar la prueba A/B. El objetivo es evitar cometer un error al realizar el cambio sin estar seguro de que el resultado sea mejor que la situación actual.

Al comenzar este tipo de prueba, siempre debemos partir de dos hipótesis: la hipótesis nula (**$H_0$**), que se trata de anular o rechazar a favor de la hipótesis alternativa (**$H_1$**). Pero, ¿cuáles serían mis hipótesis en este caso?:

- **$H_0$**: La nueva versión de la campaña no es mejor o incluso es peor que la versión anterior.

- **$H_1$**: La nueva versión de la campaña es mejor que la versión anterior.

## Dividir a los usuarios en control y test

Es fundamental dividir a los usuarios en grupos de control y test para garantizar la validez y la fiabilidad de un test A/B. Al mantener un grupo de control sin cambios, podemos aislar el efecto de la variable que estamos probando (por ejemplo, la adición de emojis en mensajes) y evaluar su impacto de manera precisa. Además, la comparación entre el grupo de control y el grupo de test nos permite determinar si el cambio implementado produce una diferencia significativa en el resultado deseado (por ejemplo, la tasa de clics en los mensajes).

Lo siguiente que nos podemos preguntar, es como debemos dividir a los usuarios en control y test, si bien no existe una norma que siempre podamos usar, algunos de las normas que debemos tener en cuenta son: 

- **Aleatorización:** La aleatorización asegura que los grupos sean seleccionados de manera imparcial, minimizando el sesgo y permitiendo la inferencia causal sobre el efecto del cambio.

- **Unidad de aleatorización:** La unidad de aleatorización, que en este caso son los usuarios individuales, garantiza que cada usuario tenga la misma probabilidad de ser asignado al grupo de control o al grupo de test, evitando sesgos de selección.

- **Proporción de división:** La proporción en la que dividimos a los usuarios entre los grupos debe ser cuidadosamente considerada para equilibrar la eficiencia de la prueba y la validez de los resultados. Podríamos tomar la decisión de coger nuestros usuarios y dividirlos aleatoriamente en 2 grupos, haciendo una división de 50/50. Los usuarios son la unidad de aleatorización, lo que garantizará que los grupos sean exactamente iguales excepto por el tratamiento. Pero cuidado, también podríamos hacer una división desigual, como 80/20, pero en este caso podríamos no tener suficientes datos para alcanzar conclusiones robustas o incluso podríamos sesgar nuestros resultados, lo que llevaría a resultados poco fiables.

## Análisis de potencia de la prueba

Para entender este punto en el flujo de trabajo debemos debemos dar un paso para atrás. Cuando nos enfrentamos a este tipo de pruebas, es normal encontrarse con 2 tipos de errores que ya conocemos y que debemos decidir un umbral que estamos dispuestos a aceptar de estos errores. Hablemos de cada uno de estos errores: 

- Nivel de significancia (Error Tipo I): Recordemos que es la probabilidad de rechazar la $H_0$ cuando realidad es cierta. Viene definido por la siguiente fórmula: 

    $α = P(rechazar H_0 | H_0 cierta)$

    En este caso, querríamos mantener este error lo más bajo posible. El **valor usado normalmente es 0.05 o 5%**, lo que se traduce como, 95 de cada 100 veces no rechazaremos la hipótesis nula cuando sea verdadera.

- Potencia estadística (Error Tipo II): Comenteremos este error al aceptar la hipótesis nula siendo falsa. Viene definida por la siguiente fórmula: 

    $β = P(aceptar H_0 | H_0 falsa)$

    Al igual que antes, querremos que este error sea lo más bajo posible, lo que haría que la potencia estadística fuera alta. El **valor normalmente usado es 0.8 u 80%**.

Pero... ¿todo esto para qué sirve? La respuesta es que nos ayudará a calcular el tamaño mínimo de la muestra necesario. Y para esto podemos usar la siguiente fórmula: 

n = $\dfrac{2 \cdot (Z_{\alpha/2} + Z_{\beta})^2 \cdot \sigma^2}{\delta^2}$

Donde: 

- $n$: Es el tamaño de la muestra por grupo que queremos calcular. 

- $\delta$: Es la diferencia mínima detectable entre la métrica elegida (por ejemplo, tasa de conversión) en los dos grupos.

-  $Z_{\alpha/2}$: Es el valor Z correspondiente al nivel de significancia deseado (por ejemplo, 0.05).

- $Z_{\beta}$: Es el valor Z correspondiente a la potencia deseada del test (por ejemplo, 0.80).

- σ: estimación de desviación entándar en el grupo de control. Esto se puede estimar a partir de datos históricos. 


Para uno de nuestros ejemplos (cambios en la página de inicio del sitio web, donde mediamos la tasa de conversión), supongamos que la tasa de conversión diaria promedio de los últimos 6 meses (sin haber cambiado nada en la página web) es de 0.15 y la desviación estándar es de 0.05. Con la nueva funcionalidad, esperamos ver un aumento absoluto del 3% en la tasa de conversión, es decir nuestra diferencia mínima detectable ($\delta$). Por lo tanto, la tasa de conversión para el grupo de tratamiento será de 0.18. También, asumimos que las desviaciones estándar muestrales son iguales para los dos grupos (σ). 

Ahora solo nos quedan dos parámetros. La $Z_{\alpha/2}$, cuyo valor comúnmente usado es de 0.05. La $Z_{\beta}$, cuyo valor normalmente usado es de 0.20, lo que corresponde a una potencia del 80%. Esto significa que tenemos una probabilidad del 80% de detectar un efecto real si realmente existe. 


Todo esto es esencial para calcular un buen tamaño de muestra y garantizar que los resultados del estudio sean representativos y generalizables. Si lo pensamos logicamente, a medida que aumenta el tamaño de la muestra, también aumenta la potencia de la prueba. La potencia de una prueba representa la probabilidad de aceptar una $H_1$ (hay diferencia entre variación A y variación B). Python tiene una librería que nos va a permitir calcular el tamaño muestral de forma sencilla, veamos como hacerlo:


````python
# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd
import numpy as np

# Calcular el tamaño muestral
# -----------------------------------------------------------------------
from statsmodels.stats.power import NormalIndPower
import statsmodels.stats.api as sms


# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns


# Importamos las funciones del archivo de soporte
# ------------------------------------------------------------------------------
from src import soporte_abtesting as sa


# Otros recursos
# ------------------------------------------------------------------------------
import math

# Configuración de las gráficas
# ------------------------------------------------------------------------------
# establece el estilo del fondo de las gráficas como "whitegrid". Las gráficas tendrán un fondo blanco con líneas de cuadrícula horizontales y verticales.
sns.set_style("dark")

# cambiará el esquema de colores predeterminado de Seaborn a uno más oscuro, adecuado para gráficas con fondos claros. 
sns.set_theme(style="dark")

# establece el estilo en seaborn-v0_8-darkgrid. 
plt.style.use('dark_background')
````

````python
# los estilos que tenemos en sns.set_theme son:
# darkgrid, whitegrid, dark, white, ticks

# los estilos que tenemos para sns.set_style son: 
# darkgrid, whitegrid, dark, white, ticks

# para ver los estilos de plt.style decomentar la linea de código 
#plt.style.available
````

````python
# calculamos la diferencia esperada entre las tasas de conversión del diseño viejo y el nuevo, que es una medida estandarizada de la diferencia entre dos proporciones.
diferencia = sms.proportion_effectsize(0.13, 0.15)  

# definimos una variable para establecer el nivel de significancia
alpha = 0.05 

# definimos una variable para establecer el la potencia deseada
power = 0.80 

# usamos el método NormalIndPower, para poder aplicar la fórmula que hemos aprendido
analisis = NormalIndPower()

# aplicamos el método 'solve_power' para calcular el tamaño muestral. Este método no hace otra cosa más que aplicar la fórmula que vimos previamente
tamaño_nuestral = analisis.solve_power(diferencia, power = power, nobs1= None, ratio = 1.0, alpha = alpha)
print(f'El tamaño muestral debe de ser de: {math.ceil(tamaño_nuestral)}')
````

# Tipos de Pruebas en AB Testing

Como hemos visto, el AB testing es una técnica utilizada para comparar dos o más variantes de un elemento con el objetivo de determinar cuál produce un mejor rendimiento en términos de una métrica específica. Sin embargo, elegir la prueba estadística adecuada es fundamental para obtener conclusiones válidas y significativas. En AB testing, las pruebas se pueden clasificar en dos categorías principales: paramétricas y no paramétricas.

- **Pruebas Paramétricas**: Asumen que los datos siguen una distribución específica (por lo general, una distribución normal) y están influenciadas por parámetros específicos. Estas pruebas son adecuadas cuando se cumplen ciertas asunciones sobre la distribución de los datos, como la homogeneidad de las varianzas y la normalidad. Las pruebas parámetricas más comunes son: 

    - **Prueba t de Student:** Se utiliza para comparar las medias de dos muestras independientes.

    - **Dependent t-Test**: Se utiliza para comparar las medias de dos muestras independientes cuando los datos siguen una distribución normal.
    
    - **Análisis de Varianza (ANOVA):** Útil para comparar las medias de más de dos grupos.

    - **Z-Test**: Utilizado para comparar la media de una muestra con la media poblacional conocida cuando el tamaño de la muestra es grande.

-  **Pruebas No Paramétricas**: Las pruebas no paramétricas son menos restrictivas en términos de asunciones sobre la distribución de los datos. Estas pruebas son útiles cuando los datos no siguen una distribución normal o cuando no se pueden cumplir los supuestos requeridos por las pruebas paramétricas. Las principales pruebas no paramétricas son: 

    - **Prueba U de Mann-Whitney:** Se emplea para comparar dos muestras independientes cuando los datos no siguen una distribución normal.

    - **Prueba de Wilcoxon:** Similar a la prueba t, pero se utiliza cuando los datos no son normales o cuando la muestra es pequeña.

    - **Prueba de Kruskal-Wallis**: Se utiliza para comparar las medianas de más de dos grupos independientes.
    
    - **Prueba de Chi-cuadrado**: Se utiliza para determinar si existe una asociación significativa entre dos variables categóricas.


# Asunciones

Como hemos comentado, en las pruebas parámetricas tenemos algunas asunciones sobre los datos que deben cumplirse para que el test sea válido. Estas asunciones son:

- **Independencia**: Las observaciones dentro de cada grupo son independientes entre sí. Esta asunción se verifica durante el diseño del estudio. Significa que cada observación/dato es independiente de otra; si se están comparando 2 o más grupos, entonces se refiere al hecho de que los grupos son mutuamente excluyentes, es decir, cada individuo pertenece solo a 1 grupo; y que los datos no se repiten en el tiempo.

- **Normalidad**: La distribución de cada grupo sigue una distribución normal. 

- **Homogeneidad de varianza (homocedasticidad)**: La varianza de las observaciones en cada grupo es aproximadamente igual. 

- **Muestreo aleatorio**: Los datos se obtienen a través de un proceso de muestreo aleatorio de la población, igual que en la asunción de la independencia, tendremos que asegurarnos que esta asunción se cumple durante el diseño del estudio. 

La violación de estas asunciones pueden afectar la validez del test elegido y llevar a conclusiones incorrectas. 

## Normalidad

Como hemos dicho, muchas pruebas paramétricas se basan en esta premisa. Implica que la distribución de los datos sigue una distribución normal o gaussiana, lo que significa que los datos están simétricamente distribuidos alrededor de su media, con la mayoría de los valores cerca de la media y pocos valores en los extremos. Es importante destacar que esta asunción solo la tendremos que testar si nuestra variable dependiente es contínua. 

Esta asunción se basa en el Teorema Central del Límite (TCL), que establece que, bajo ciertas condiciones, la distribución de la media de una muestra tiende a seguir una distribución normal a medida que el tamaño de la muestra aumenta. **Esta propiedad del TCL es fundamental para muchos métodos estadísticos paramétricos, como la prueba t, el análisis de varianza (ANOVA) y la regresión lineal**.


Para poder identificar si nuestros datos presentan una distribución normal tenemos distintas aproximaciones: 

- **Aproximación visual**: En Python, hay varias herramientas visuales que se pueden utilizar para evaluar la normalidad de una distribución de datos. Algunas de las visualizaciones más comunes incluyen:

  - **Histogramas**: Los histogramas son útiles para visualizar la distribución de frecuencia de los datos. Al observar la forma del histograma, se pueden identificar indicios de normalidad, como una forma de campana simétrica.

  - **Gráficos Q-Q (Quantile-Quantile)**: Los gráficos Q-Q comparan los cuantiles teóricos de una distribución (como la normal) con los cuantiles observados de los datos. Si los puntos en el gráfico se alinean aproximadamente en una línea recta, indica que los datos se ajustan a la distribución especificada.


- **Aproximación análitica**: En Pyhton tenemos una serie de métodos que nos van a permitir evaluar si nuestros datos siguen una distribución normal.

  -  **Test de Shapiro-Wilk**: Calcula una estadística W y evalúa si la muestra de datos proviene de una distribución normal.

      - Se utiliza principalmente cuando se tiene una muestra de datos relativamente pequeña (generalmente menos de 50 observaciones).

      - Hipótesis nula (H0): Los datos siguen una distribución normal.

      - Hipótesis alternativa (H1): Los datos no siguen una distribución normal.

        Su sintaxis básica es: 
          ```python
          from scipy.stats import shapiro

          # Sintaxis básica del test de Shapiro-Wilk
          statistic, p_value = shapiro(data)
          ```

          Donde:

          - `shapiro`: Es la función que realiza el test de Shapiro-Wilk.

          - `data`: Es la muestra de datos que deseas evaluar para la normalidad.

          La función `shapiro` devuelve dos valores:

          - `statistic`: El valor estadístico del test de Shapiro-Wilk, que se utiliza para evaluar si los datos siguen una distribución normal. Cuanto mayor sea este valor, más probable es que los datos sean normales.

          - `p_value`: El valor p (p-value) asociado al test de Shapiro-Wilk, que indica cuán significativa es la desviación de la normalidad. Un valor p pequeño sugiere que los datos no siguen una distribución normal.


  -  **Test de Kolmogorov-Smirnov**: Compara la distribución de los datos con una distribución normal.

      - Se utiliza cuando se tienen muestras más grandes, aunque también puede aplicarse a muestras pequeñas (pero es menos sensible a detectar falta de normalidad en estos casos)

      - Hipótesis nula (H0): Los datos provienen de una distribución normal.

      - Hipótesis alternativa (H1): Los datos no provienen de una distribución normal.

        Su sintaxis básica es: 

          ```python
          from scipy.stats import levene

          # Sintaxis básica del test de Levene
          statistic, p_value = levene(group1, group2, group3, ...)
          ```

          Donde:

          - `levene`: Es la función que realiza el test de Levene.

          - `group1`, `group2`, `group3`, ...:  Son los grupos de datos que deseas comparar. Es decir, cada uno de los valores únicos que tiene la columna (en nuestro caso Control Campaign y Test Campaign)

          La función `levene` devuelve dos valores:

          - `statistic`: El valor estadístico del test de Levene, que se utiliza para evaluar la homogeneidad de varianza.

          - `p_value`: El valor p (p-value) asociado al test de Levene, que indica cuán significativas son las diferencias en las varianzas entre los grupos.

  En ambos test, la interpretación de los resultados es la misma, ya que nos fijaremos en el p-valor que nos devuelve los métodos: 

    - Si el p-value > 0.05, lo que sugiere que los datos siguen una distribución normal.

    - Si el p-value < 0.05, lo que sugiere que los datos no siguen una distribución normal.



## Homogeneidad de Varianzas (Homocedasticidad)

Lo que se persigue con esta asunción es conseguir la igualdad de varianzas a lo largo de diferentes grupos o niveles de una variable independiente. En otras palabras, implica que las diferencias entre los valores del grupo control y los valores del grupo test son similares y no dependen de los valores de la variable independiente. Esta asunción es especialmente importante en modelos de regresión lineal y análisis de varianza (ANOVA). Tenemos varias formas de poder evluar la homocedasticidad de nuestros datos: 


  - **Test de Levene**: Evalúa si las varianzas de dos o más grupos de datos son iguales. Es menos sensible a las desviaciones de la normalidad en comparación con otros tests como el de Bartlett.

      - Se utiliza cuando los datos no cumplen con la suposición de normalidad.

      - Hipótesis nula (H0): Las varianzas de los grupos son iguales.

      - Hipótesis alternativa (H1): Las varianzas de los grupos son diferentes.

  - **Test de Bartlett**: También evalúa la igualdad de varianzas entre diferentes grupos de datos.

      - Se utiliza cuando los datos siguen una distribución normal.

      - Hipótesis nula (H0): Las varianzas de los grupos son iguales.

      - Hipótesis alternativa (H1): Las varianzas de los grupos son diferentes.


  Al igual que en los test de normalidad, para decidir si se cumple o no esta asunción deberemos fijarnos en el p valor del test. En este sentido, y al igual que en los test de normalidad: 

  - Si el valor p resultante es mayor que el nivel de significancia (generalmente 0.05), no rechazamos la hipótesis nula y concluimos que los datos pueden provenir de una distribución normal.

  - Si el valor p es menor que el nivel de significancia, rechazamos la hipótesis nula y concluimos que los datos no siguen una distribución normal.

## Z-test

En nuestro experimento estamos evaluando si el cambio en nuestra web nos ayuda a tener una mayor tasa de conversión. Para eso,  estamos mostrando al grupo control el diseño antiguo, mientras que al grupo test les mostramos el nuevo diseño. Además, hemos medido si cada uno de los usuarios a convertido en compra, asignando 1 a los que si convirtieron en compra y 0 a los que no. 

Otros escenarios en los que podemos usar este tipo de aproximación son: 

- **Comparación de tasas de conversión en dos variantes de una página web:** Es el ejemplo que estamos viendo en la clase. Se utiliza para comparar la tasa de conversión de la versión actual de una página web con una nueva variante. 

- **Comparación de tasas de clics en dos anuncios publicitarios:**  Se emplea para evaluar si hay una diferencia significativa en la tasa de clics entre dos anuncios publicitarios diferentes.


El Z-test nos va a permitir comparar las **proporciones de dos grupos** y determinar si hay una diferencia significativa entre ellas. En nuestro caso, el Z-test lo vamos a utilizar para comparar la tasa de conversión entre dos grupos: el grupo control y el grupo de test. Nos va a ayudar a determinar si la diferencia en las tasas de conversión entre estos grupos es estadísticamente significativa o simplemente debido a la variabilidad aleatoria en los datos, es decir, por azar. 


El Z-score se calcula utilizando la siguiente fórmula:

$Z = \frac{(p_1 - p_2)}{\sqrt{\frac{p_1(1-p_1)}{n_1} + \frac{p_2(1-p_2)}{n_2}}}$

- $p_1$ y $p_2$: Representan las proporciones de éxito en los grupos 1 y 2, respectivamente (las tasas de conversión de cada grupo).

- $n_1$ y $n_2$: Son los tamaños de las muestras en los grupos 1 y 2, respectivamente.

Como siempre, en Python, tenemos un método que nos va a permitir aplicar esta fórmula fácilmente. El método que deberemos usar es el `proportions_ztest` de la librería `statsmodels`. Su sintaxis básica es: 

```python
from statsmodels.stats.proportion import proportions_ztest

proportions_ztest(count, nobs, alternative='two-sided')
```

Donde: 

- `count`: Una lista o array que contiene el número de éxitos en cada grupo. Por ejemplo, si estamos comparando la tasa de éxito entre dos grupos A y B, `count` podría ser `[successes_A, successes_B]`.

- `nobs`: Una lista o array que contiene el tamaño  muestral en cada grupo. Por ejemplo, si tenemos dos grupos A y B, `nobs` podría ser `[25, 25]`.

- `alternative` (opcional): Especifica la hipótesis alternativa y puede ser `'two-sided'`, `'smaller'` o `'larger'`. Por defecto, es `'two-sided'`, lo que significa que estamos interesados en si hay una diferencia en cualquier dirección entre las proporciones de los dos grupos.


**Los resultados que obtenemos de este método son**:

- **Estadístico de Prueba Z ($Z$):** Representa cuántas desviaciones estándar está una observación de la media muestral en relación con la hipótesis nula. Pero que quiere decir esto: 

  - Un $Z$ positivo indica que la media muestral está por encima de la media poblacional asumida. 
  
  - Un $Z$ negativo indica que está por debajo. 
  
  - Cuanto mayor sea el valor absoluto de $Z$, más evidencia hay en contra de la hipótesis nula, es decir, que no hay diferencia entre los grupos. 

- **p-value:** Es el valor que realmente nos va a ayudar a decidir si hay evidencias suficientes para decir si hay diferencias entre los grupos. Representa la probabilidad de obtener un estadístico de prueba $Z$ al menos tan extremo como el observado, asumiendo que la hipótesis nula es verdadera.

**Interpretación:**

- Si `p_value < 0.05`: Rechazamos la hipótesis nula. Hay evidencia suficiente para decir que hay una diferencia significativa entre las dos muestras.

- Si `p_value >= 0.05`: No hay evidencia suficiente para rechazar la hipótesis nula. No podemos afirmar que haya una diferencia significativa entre las dos muestras.

````python
# instanciamos la clase de los métodos estadísticos
ab_testing = sa.Pruebas_parametricas("group",  "converted", ab_test, "treatment", "control")
ab_testing.z_test()
````

## Análisis de Varianza (ANOVA)

El ANOVA es una técnica estadística utilizada para comparar las medias de dos o más grupos para determinar si al menos uno de los grupos difiere significativamente de los demás.

Demos un contexto para entender mejor cuando usar este tipo de aproximación. Supongamos que somos analistas de marketing digital y estamos probando tres versiones diferentes de una página web para ver cuál tiene el mayor impacto en el tiempo promedio que los usuarios pasan en la página. Tenemos tres grupos: el Grupo A ve la versión actual, el Grupo B ve una variación con cambios en el diseño, y el Grupo C ve otra variación con cambios en el contenido.

En este contexto, dado que tenemos tres grupos que comparar tendremos que usar el ANOVA, siempre y cuando los datos cumplan las asunciones de normalidad, independencia y homocedasticidad. Pero recordemos, que no podemos enfrentarnos a un problema de A/B testing sin definir cuáles son nuestras hipótesis. En este caso: 

- **Hipótesis Nula ($H0$):** No hay diferencia significativa en el tiempo promedio que los usuarios pasan en la página entre los grupos.

- **Hipótesis Alternativa ($H1$):** Hay una diferencia significativa en el tiempo promedio que los usuarios pasan en la página entre los grupos.

Otros ejemplos donde podemos usar este tipo de aproximación son: 

- **Comparación de rendimiento entre equipos de ventas por región**: La empresa quiere determinar si hay diferencias en las ventas entre equipos de ventas en diferentes regiones. Se aplica ANOVA para comparar las medias de ventas entre las regiones.

- **Análisis de satisfacción del cliente en diferentes canales de atención**: La empresa examina si hay variaciones significativas en la satisfacción del cliente entre diferentes canales de atención, como chat en vivo, correo electrónico y teléfono. ANOVA se utiliza para comparar las medias de satisfacción.

La idea detrás del ANOVA es comparar la variabilidad entre grupos (SSG) con la variabilidad dentro de los grupos (SSE) para determinar si las diferencias entre las medias de los grupos son estadísticamente significativas. Si la variabilidad entre grupos es significativamente mayor que la variabilidad dentro de los grupos, entonces podemos concluir que al menos un grupo difiere de los demás en términos de la variable que estamos estudiando. Para poder trabajar con esta aproximación debemos calcular la estadística de prueba F, la cual viene definida por la siguiente fórmula: 

$F = \frac{{\text{Variación entre grupos}}}{{\text{Variación dentro de los grupos}}}$

La siguiente pregunta que nos podemos hacer es, ¿como podemos calcular la variación entre grupos y la variación dentro de los grupos? Para calcular cada elemento de la fórmula

1. **Suma de cuadrados entre grupos (SSG, variación entre grupos) :** Esta medida cuantifica la variabilidad entre las medias de los diferentes grupos. Se calcula sumando las diferencias al cuadrado entre la media de cada grupo y la media global, y luego multiplicando esa diferencia por el número de observaciones en cada grupo. Es decir, se mide cuánto difieren las medias de los grupos entre sí.

2. **Suma de cuadrados dentro de los grupos (SSE, variación detro de los grupos):** Esta medida cuantifica la variabilidad dentro de cada grupo, es decir, la variabilidad que no se explica por las diferencias entre las medias de los grupos. Se calcula sumando las diferencias al cuadrado entre cada observación y la media de su propio grupo, y luego sumando estos valores para todos los grupos. Es decir, mide cuánto varían las observaciones dentro de cada grupo.


Para realizar ANOVA en Python, podemos utilizar la función `f_oneway` de la librería `scipy.stats`, que se utiliza para comparar las medias de tres o más grupos. Su sintaxis básica es: 

```python
from scipy import stats

# Aplicar ANOVA
resultado_anova = stats.f_oneway(grupo_1, grupo_2, grupo_3)

# Imprimir resultados
print("Estadístico F:", resultado_anova.statistic)
print("Valor p:", resultado_anova.pvalue)
```

- `grupo_1`, `grupo_2`, `grupo_3`: Listas o arrays que representan los datos de cada grupo.

- `resultado_anova`: Objeto que contiene el resultado del análisis de varianza.

- `resultado_anova.statistic`: Estadístico F.

- `resultado_anova.pvalue`: Valor p asociado al estadístico F.



**Los resultados que obtenemos son:**

- **Estadístico F:** El estadístico F es el resultado principal de ANOVA. Representa la relación entre la variabilidad entre los grupos y la variabilidad dentro de los grupos.  Cuando el valor de F es grande y el valor p asociado es pequeño, indica que existe una diferencia significativa entre al menos dos grupos en estudio. Entendámoslo un poco más, imaginemos que estamos realizando un ANOVA para comparar las calificaciones promedio de tres métodos de enseñanza diferentes: A, B y C. 

    - Si el valor de F es grande y el valor p es pequeño (por ejemplo, menor que 0.05), entonces esto sugiere que al menos uno de los métodos de enseñanza tiene un efecto significativo en las calificaciones promedio de los estudiantes.

    - En este contexto, si el valor de F es alto, significa que la variabilidad entre las medias de los grupos (es decir, las calificaciones promedio de los tres métodos de enseñanza) es mucho mayor que la variabilidad dentro de cada grupo (es decir, las diferencias individuales en las calificaciones dentro de cada método de enseñanza). Esto sugiere que las diferencias entre los métodos de enseñanza son lo suficientemente grandes como para no ser atribuibles únicamente al azar o la variabilidad inherente de los datos.


- **Valor p:** El valor p asociado al estadístico F indica la probabilidad de obtener un valor de estadístico F tan extremo como el observado, asumiendo que las medias de todos los grupos son iguales. Un valor p pequeño (< 0.05) sugiere que puedes rechazar la hipótesis nula de igualdad de medias.

**Interpretación:**

- Si `p_value < 0.05`: Rechazamos la hipótesis nula. Hay evidencia suficiente para decir que hay una diferencia significativa entre las dos muestras.

- Si `p_value >= 0.05`: No hay evidencia suficiente para rechazar la hipótesis nula. No podemos afirmar que haya una diferencia significativa entre las dos muestras.

````python
# generamos los datos ficticios del Grupo A (versión actual)
grupo_a = np.random.normal(60, 10, 100)

# generamos los datos ficticion del Grupo B (variación de diseño)
grupo_b = np.random.normal(65, 10, 100)

# generamos los datos ficticios del Grupo C (variación de contenido)
grupo_c = np.random.normal(55, 10, 100)

# Crear DataFrame
df_anova = pd.DataFrame({
    'grupo': ['A'] * 100 + ['B'] * 100 + ['C'] * 100,
    'tiempo_promedio': np.concatenate([grupo_a, grupo_b, grupo_c])
})

# Visualizar los primeros registros del DataFrame
df_anova.sample(6)
````

## t-Student

La t de Student se utiliza para realizar comparaciones entre dos grupos y determinar si hay una diferencia significativa entre ellos.

Como hasta ahora, demos contexto al ejemplo para entender mejor. Supongamos que somos analistas de marketing digital que trabajamos para una empresa de comercio electrónico. Recientemente, hemos realizado una actualización en el diseño de los anuncios de la web, y ahora estamos interesados en evaluar si la nueva versión de los anuncios tiene un impacto significativo en la tasa de clics (CTR) en comparación con la versión anterior.

Para realizar esta evaluación, vamos a implementar un A/B test, donde hemos dividido aleatoriamente a los usuarios en dos grupos: el grupo de control, que ve la versión antigua del anuncio, y el grupo de prueba, que ve la nueva versión del anuncio. Cada grupo está compuesto por 1000 usuarios. Las hipótesis que hemos definido son: 

- **Hipótesis Nula ($H0$):** No hay diferencia significativa en las tasas de clics entre el grupo de control y el grupo de prueba.

- **Hipótesis Alternativa ($H1$):** Hay una diferencia significativa en las tasas de clics entre el grupo de control y el grupo de prueba.

Otros escenarios donde podemos usar la t-Student son: 

- **Comparación de rendimiento entre dos equipos de ventas:** Supongamos que una empresa quiere determinar si hay una diferencia significativa en el rendimiento de ventas entre dos equipos de ventas: el equipo A y el equipo B. Se recopila una muestra de las ventas mensuales de cada equipo durante un período de tiempo específico. Utilizando la prueba t de Student, la empresa puede comparar las medias de ventas mensuales de ambos equipos para determinar si hay una diferencia significativa en el rendimiento de ventas entre ellos.

- **Evaluación de la efectividad de dos estrategias de marketing:** Imagina que una empresa implementa dos estrategias de marketing diferentes, A y B, para promocionar un producto nuevo. Después de un período de tiempo, se recopilan datos sobre las ventas generadas por cada estrategia. La empresa puede utilizar la prueba t de Student para comparar las medias de ventas entre las dos estrategias y determinar si hay una diferencia significativa en la efectividad de marketing entre ellas.


La prueba t nos proporciona un valor t que nos dice cuántas desviaciones estándar está la diferencia de medias por encima o por debajo de cero. Si este valor es grande, sugiere que la diferencia es significativa y no simplemente el resultado del azar.

Como hemos dicho, la prueba t de Student para A/B testing implica calcular un valor t que luego se compara con un valor crítico de la distribución t para determinar si la diferencia entre las medias de dos muestras es estadísticamente significativa. La fórmula para calcular el valor t en este contexto es:

$t = \frac{(\bar{x}_1 - \bar{x}_2)}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$

Donde:
- $\bar{x}_1$ y $\bar{x}_2$: Son las medias de las muestras 1 y 2 respectivamente.

- $s_1$ y $s_2$: Son las desviaciones estándar de las muestras 1 y 2 respectivamente.

- $n_1$ y $n_2$: Son los tamaños de las muestras 1 y 2 respectivamente.


Para realizar esta prueba usaremos el método `ttest_ind`. Su sintaxis básica :

```python
from scipy import stats

t_stat, p_value = ttest_ind(a, b, equal_var=True, alternative='two-sided')
```

Donde: 
- `a` y `b`: Son las dos muestras que queremos comparar.

- `equal_var` (opcional): Es un parámetro opcional que indica si se debe asumir igualdad de varianzas. Por defecto es True.

- `alternative` (opcional): Indica la hipótesis alternativa ('two-sided', 'less', o 'greater'). Por defecto es 'two-sided'. Entendemos cada uno de estos valores que puede recibir este parámetro: 

    - **'two-sided'**: También conocido como prueba de dos colas. En esta hipótesis alternativa, estamos interesados en si hay una diferencia significativa entre las medias de los dos grupos, sin importar si una es mayor o menor que la otra. Es decir, estamos buscando cualquier tipo de diferencia, ya sea positiva o negativa.

    - **'less'**: Esta opción se usa cuando estamos interesados en probar si la media del primer grupo es significativamente menor que la del segundo grupo. Por ejemplo, si estamos comparando dos versiones de un sitio web y queremos ver si la primera versión tiene menos clics que la segunda.

    - **'greater'**: Se utiliza cuando queremos probar si la media del primer grupo es significativamente mayor que la del segundo grupo. Por ejemplo, si estamos comparando dos tratamientos médicos y queremos ver si el primero tiene una tasa de éxito mayor que el segundo.



**Los resultados que obtenemos son**:

- **t_stat:** Este es el valor estadístico de la prueba t de Student. Cuanto mayor sea este valor (en términos absolutos), más evidencia hay en contra de la hipótesis nula.

- **p_value:** Este es el valor p de la prueba. Representa la probabilidad de obtener un valor estadístico tan extremo como el observado, bajo la hipótesis nula. 

**Interpretación:**

- Si `p_value < 0.05`: Rechazamos la hipótesis nula. Hay evidencia suficiente para decir que hay una diferencia significativa entre las dos muestras.

- Si `p_value >= 0.05`: No hay evidencia suficiente para rechazar la hipótesis nula. No podemos afirmar que haya una diferencia significativa entre las dos muestras.

````python
# crear un diccionario con dos columnas, la del grupo a la que pertecen y otra con los valores de CTR (con una distribución normal)
data_t = {'Grupo': ['Control'] * 25 + ['Test'] * 25,
        'CTR': np.concatenate([np.random.normal(0.02, 0.01, 25),
                               np.random.normal(0.025, 0.01, 25)])}

df_t = pd.DataFrame(data_t)
df_t.head()
````

## t-Student dependiente

Esta prueba, también conocida como prueba t de muestras pareadas, se utiliza para comparar las medias de dos muestras relacionadas o dependientes. Esto es útil cuando se tienen dos conjuntos de datos que están emparejados de alguna manera, es decir, los sujetos de la pruena contribuyen en los dos grupos. 


Como hasta ahora demos un contexto para entender mejor esta aproximación. Un gimnasio quiere evaluar la efectividad de su nuevo programa de entrenamiento físico en la mejora de la fuerza muscular de sus miembros. Para ello, se ha diseñado un estudio donde se mide la fuerza muscular de un grupo de participantes antes y después de completar el programa de entrenamiento de 8 semanas. Las hipótesis que planteamos en este problema son: 

- Hipótesis nula ($H0$): No hay diferencia en la fuerza muscular antes y después del programa. 

- Hipótesis alternativa ($H1$): Hay una diferencia en la fuerza muscular antes y después del programa. 

Otros ejemplos donde podemos aplicar esta aproximación son: 

- **Efectividad de un nuevo método de estudio:** Una universidad evalúa un nuevo método de estudio midiendo las calificaciones de los estudiantes antes y después de su implementación. Comparan las medias usando una prueba t de Student dependiente para determinar si el método mejora significativamente las calificaciones.

- **Evaluación del rendimiento de un programa de capacitación laboral:**  Una empresa implementa un programa de capacitación para mejorar las habilidades de sus empleados en un determinado campo. Miden las puntuaciones de las pruebas de competencia antes y después del programa. Utilizan una prueba t de Student dependiente para determinar si el programa produce mejoras significativas en las habilidades de los empleados.

La idea detrás de la prueba t de Student dependiente es comparar las medias de dos muestras relacionadas o emparejadas. De la misma forma que con la t de Student, en esta aproximacion se calcula el valor de t, la cual nos indica cuántas desviaciones estándar está la diferencia entre las medias de las dos muestras relacionadas por encima o por debajo de cero. Un valor t grande sugiere una diferencia significativa, no simplemente aleatoria.


En la prueba t de Student para muestras dependientes, calculamos un valor t que se compara con un valor crítico de la distribución t para determinar la significancia estadística de la diferencia entre las medias de dos muestras relacionadas. La fórmula para calcular el valor t es:

$t = \frac{(\bar{x}_1 - \bar{x}_2)}{\sqrt{\frac{s_d^2}{n}}}$

Donde:
- $\bar{x}_1$ y $\bar{x}_2$: Representan las medias de las dos muestras relacionadas.

- $s_d$: Es la desviación estándar de las diferencias entre las dos muestras.

- $n$: Es el tamaño de la muestra.

Para realizar esta prueba, usamos el método `ttest_rel`. Su sintaxis básica es:

```python
from scipy import stats

 t_stat, p_value = stats.ttest_rel(a, b, alternative = "two-sided", nan_policy = "propagate")
```

Donde:
- `a` y `b`: Son las dos muestras relacionadas que queremos comparar.

- `alternative`: Especifica la hipótesis alternativa. Puede tomar uno de los siguientes valores:

    - `'two-sided'` (predeterminado): Significa que estamos interesados en si hay una diferencia significativa en cualquier dirección.

    - `'less'`: Significa que estamos interesados en si la diferencia es significativamente menor que cero.

    - `'greater'`: Significa que estamos interesados en si la diferencia es significativamente mayor que cero.

- `nan_policy`: Especifica cómo manejar los valores nulos. Puede tomar los siguientes valores: 

    - `'propagate'`: Si se encuentran valores NaN en las muestras, la función devolverá NaN.
    
    - `'raise'`: Si se encuentran valores NaN en las muestras, la función generará un error `ValueError`.

    - `'omit'`: Si se encuentran valores NaN en las muestras, se omitirán y la función calculará el resultado sin ellos.

    - `'pairwise'`: Si se encuentran valores NaN en las muestras, se manejarán de manera "pareada". En otras palabras, se ignorarán los valores NaN en cada par de observaciones al calcular la estadística de la prueba.


**Los resultados que obtenemos son:**

- `t_stat`: Es el valor estadístico calculado para la prueba t de Student. Indica cuántas desviaciones estándar está la diferencia de medias de las muestras por encima o por debajo de cero.

- `p_value`: Es el valor de probabilidad asociado al estadístico t. Representa la probabilidad de obtener un valor estadístico tan extremo como el observado, asumiendo que la hipótesis nula es verdadera.

**Interpretación:**

- Si `p_value < 0.05`: Rechazamos la hipótesis nula. Hay evidencia suficiente para decir que hay una diferencia significativa entre las dos muestras.

- Si `p_value >= 0.05`: No hay evidencia suficiente para rechazar la hipótesis nula. No podemos afirmar que haya una diferencia significativa entre las dos muestras.

# Elección de la métrica principal para la prueba A/B

En este jupyter hemos visto una métrica para realizar nuestro análisis. Pero la realidad es que elegir la métrica con la que vamos a querer trabajar es una de las partes más importantes de la prueba A/B, ya que esta métrica se utilizará para medir el rendimiento del producto para los grupos test y control, y se utilizará para identificar si hay una diferencia estadísticamente significativa entre ellos. La elección de la métrica depende de la hipótesis subyacente que estemos probando. Esta es, si no la más importante, una de las partes más importantes de la prueba A/B porque determina cómo se diseñará la prueba y también qué tan bien funcionarán las ideas propuestas. 

Los ingresos o la tasa de conversión no siempre son el objetivo final de un problema al que nos enfrentamos, así que para la prueba A/B, necesitamos vincular la métrica principal a los objetivos directos del producto. Para poder elegir la mejor métrica, podemos hacernos la siguiente pregunta a la hora de empezar:

> Si esta métrica elegida aumentará significativamente mientras todo lo demás permanece constante, ¿lograríamos nuestro objetivo y abordaríamos el problema?


A continuación os listamos algunas de las métricas que podríamos usar segun el sector para que tengamos algunos ejemplos: 

| Sector              | Métrica                                         | Definición                                                    | Fórmula                                                       | Ejemplo                                                    |
|---------------------|-------------------------------------------------|---------------------------------------------------------------|---------------------------------------------------------------|------------------------------------------------------------|
